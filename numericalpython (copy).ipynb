3.ARRAY PROPERTIES

```python
import numpy as np
a=np.array([[1,2,3],[4,5,6]])
print(a.shape)#give the rows and columns in array
print(a.size)#gives the total elements present in array
print(a.ndim)#gives the dimension of array
print(a.dtype)#gives the data types of array
```

4.INDEXING AND SLICING

```python
import numpy as np
a = np.array([10,20,30,40,50])
print(a[0])#prints the first element
print(a[-1])#prints the last elememt
print(a[1:4])#prints the slicing elements
```

```python
import numpy as np
mat = np.array([[1,2,3],
               [4,5,6]])
print(mat[0,1])#prints the 0 column and 1st row
print(mat[:,1])#prints the entire column 1
print(mat[1,:])#prints the entire row 1
```

MATHEMATICAL OPERATIONS


```python
import numpy as np
a = np.array([1,2,3])
b = np.array([10,20,30])
print(a+b)
print(a-b)
print(a*b)
print(a**b)
print(a**2)
            
```

MATRIX OPERATIONS

```python
import numpy as np
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(np.dot(a,b))# prints dot product of matrix
print(np.transpose(a))#print the transpose of a
print(a @ b)           #prints the matrix multiplication
```

FUNCITONS

```python
import numpy as np
a = np.array([10,20,30,40])
print(a.sum()) #sum
print(a.mean())#mean
print(a.max())#max value
print(a.min())#min value
print(a.argmax())#index of max value
print(a.argmin())#index of min value
```

DOING CONDITIONS LIKE GREATER AND DIVISIBLE

```python
import numpy as np
a = np.array([1,2,5,10,15,20])
filter1 = a[a>10] #prints as the condition follows
print(filter1)

filter2=a[a%2==0]
print(filter2)
```

COPY AND VIEW

```python
import numpy as np
a = np.array([1,2,3])
view = a.view()
copy = a.copy()
a[0]=100
print(view)      # gives the view of changing attribute
print(copy)      #gives the copy of the given array
```
